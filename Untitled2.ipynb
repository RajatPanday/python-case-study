```python
import math

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

def power(a, b):
    return a ** b

def modulus(a, b):
    if b == 0:
        return "Cannot modulo by zero"
    return a % b

def calculate_factorial(n):
    if n < 0:
        return "Error: Cannot compute negative factorial"
    if not n.is_integer():
        return "Error: Factorial requires an integer"
    return math.factorial(int(n))

def calculate_square_root(n):
    if n < 0:
        return "Error: Cannot calculate square root of a negative number"
    return math.sqrt(n)

def main():
    while True:
        print("\nAdvanced Calculator")
        print("1 Addition")
        print("2 Subtraction")
        print("3 Multiplication")
        print("4 Division")
        print("5 Power")
        print("6 Modulus")
        print("7 Factorial")
        print("8 Square Root")
        print("9 Exit")

        choice = input("Enter choice: ")

        if choice == '9':
            print("Program terminated")
            break

        if choice not in [str(i) for i in range(1, 10)]:
            print("Invalid choice. Try again.")
            continue

        try:
            if choice in ['1', '2', '3', '4', '5', '6']:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))

                if choice == '1':
                    print(f"Result = {add(num1, num2)}")
                elif choice == '2':
                    print(f"Result = {subtract(num1, num2)}")
                elif choice == '3':
                    print(f"Result = {multiply(num1, num2)}")
                elif choice == '4':
                    result = divide(num1, num2)
                    if isinstance(result, str):
                        print(result)
                    else:
                        print(f"Result = {result}")
                elif choice == '5':
                    print(f"Result = {power(num1, num2)}")
                elif choice == '6':
                    result = modulus(num1, num2)
                    if isinstance(result, str):
                        print(result)
                    else:
                        print(f"Result = {result}")

            elif choice in ['7', '8']:
                num = float(input("Enter number: "))

                if choice == '7':
                    result = calculate_factorial(num)
                    if isinstance(result, str):
                        print(result)
                    else:
                        print(f"Factorial = {result}")
                elif choice == '8':
                    result = calculate_square_root(num)
                    if isinstance(result, str):
                        print(result)
                    else:
                        print(f"Square Root = {result}")

        except ValueError:
            print("Invalid numeric input. Please enter valid numbers.")

if __name__ == "__main__":
    main()
```
