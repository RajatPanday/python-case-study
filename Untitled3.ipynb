```python
def classify_temperature(temp):
    if temp < 18:
        return "Cold"
    elif 18 <= temp <= 28:
        return "Normal"
    else:
        return "Hot"

def calculate_statistics(readings):
    avg_temp = sum(readings) / len(readings)
    highest = max(readings)
    lowest = min(readings)
    return avg_temp, highest, lowest

def display_report(avg_temp, highest, lowest, counts):
    print("\nTemperature Summary")
    print(f"Average Temperature = {avg_temp}")
    print(f"Highest Temperature = {highest}")
    print(f"Lowest Temperature = {lowest}")

    print("\nCondition Count")
    print(f"Hot = {counts.get('Hot', 0)}")
    print(f"Normal = {counts.get('Normal', 0)}")
    print(f"Cold = {counts.get('Cold', 0)}")

    print("\nSuggested Action")
    actions = []
    if counts.get('Hot', 0) > 0:
        actions.append("Fan ON")
    if counts.get('Cold', 0) > 0:
        actions.append("Heater ON")

    if not actions:
        print("No action required")
    else:
        for action in actions:
            print(action)

def main():
    try:
        num_readings = int(input("Enter number of readings: "))
        if num_readings <= 0:
            print("Please enter a valid positive number.")
            return

        readings = []
        for _ in range(num_readings):
            temp = float(input("Enter temperature: "))
            readings.append(temp)

        counts = {"Hot": 0, "Normal": 0, "Cold": 0}
        for temp in readings:
            condition = classify_temperature(temp)
            counts[condition] += 1

        avg_temp, highest, lowest = calculate_statistics(readings)

        display_report(avg_temp, highest, lowest, counts)

    except ValueError:
        print("Invalid input. Please enter numeric values.")

if __name__ == "__main__":
    main()
```
